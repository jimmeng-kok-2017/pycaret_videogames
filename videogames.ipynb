# Import Relevant Libraries

```python
import pandas as pd

from pycaret.regression import *
```

# Import Dataset and Data Pre-processing

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
print("Number of duplicated rows are: ",df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

# Split into Train and Test datasets

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.25, random_state=42, shuffle=True)
```

# Train, Create, and Evaluate Model

```python
experiment = setup(train, target = 'Global_Sales', ignore_features= ['Rank', 'Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])
```

```python
compare_models()
```

```python
br = create_model('br')
```

```python
tuned_br = tune_model(br, optimize = 'MAE') #lower the better, compare with Mean above
```

```python
evaluate_model(tuned_br)
```

```python
final_br = finalize_model(tuned_br)
print(final_br)
```

# Test the Model

```python
test.drop(df.columns[[6,7,8,9]],axis=1,inplace=True)
```

```python
predictions = predict_model(final_br, data=test)
predictions.head()
```
